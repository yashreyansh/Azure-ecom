```python
import pandas as pd
```

```python
df = pd.read_csv('users.6M0xxK.2024.public.csv')
df.head(2)
```

```python
chunk_size = len(df) // 5 + ( 1 if len(df)%5 else 0)

for i in range(5):
    start  = i * chunk_size
    end = start + chunk_size
    chunk_df = df.iloc[start:end]
    chunk_df.to_csv(f'ecom_user_data/user_data_{i+5+1}.csv', index = False)
```
